Simple Linear Regresion Salary Data

```python
import pandas as pd # Mengimpor pustaka pandas dan memberi alias 'pd'
```

```python
dataset = pd.read_csv('Salary_Data.csv') # Membaca file CSV ke dalam DataFrame bernama 'dataset'
```

```python
from sklearn.model_selection import train_test_split # Mengimpor fungsi 'train_test_split' dari submodul 'sklearn.model_selection'
```

```python
from sklearn.linear_model import LinearRegression # Mengimpor kelas 'LinearRegression' dari submodul 'sklearn.linear_model'
```

```python
dataset = pd.read_csv('Salary_Data.csv') # Mengulang proses membaca file CSV untuk memuat data

# Split dataset into features and target variable
X = dataset.iloc[:, :-1].values  # Mengambil semua kolom kecuali kolom terakhir sebagai fitur (Years of Experience)
y = dataset.iloc[:, -1].values   # Mengambil kolom terakhir sebagai target (Salary)

# Memisahkan dataset menjadi fitur (X) dan target (y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)

# Membagi dataset menjadi training set dan test set
# Test size = 1/3 berarti 33% data digunakan untuk testing, dan sisanya untuk training
# random_state = 0 digunakan untuk hasil split yang sama setiap kali dijalankan
regressor = LinearRegression()
# Melatih model menggunakan training set (X_train, y_train)
regressor.fit(X_train, y_train)

# Melakukan prediksi pada test set (X_test)
y_pred = regressor.predict(X_test)

# Menampilkan hasil prediksi
print(y_pred)
```

```python
# Mengimpor pustaka visualisasi matplotlib.pyplot dan memberi alias 'plt'
import matplotlib.pyplot as plt

# Visualisasi Hasil Training Set
plt.scatter(X_train, y_train, color='red')  # Menampilkan data aktual pada training set sebagai titik merah
plt.plot(X_train, regressor.predict(X_train), color='blue') # Menampilkan garis regresi yang dihasilkan dari data training
plt.title('Salary vs Experience (Training set)') # Memberikan judul pada grafik
plt.xlabel('Years of Experience') # Memberikan label pada sumbu X
plt.ylabel('Salary') # Memberikan label pada sumbu Y
plt.show() # Menampilkan grafik

# Visualisasi Hasil Test Set
plt.scatter(X_test, y_test, color='red')  # Menampilkan data asli pada test set
plt.plot(X_train, regressor.predict(X_train), color='blue')  # Garis regresi tetap menggunakan hasil dari training set
plt.title('Salary vs Experience (Test set)') # Memberikan judul pada grafik
plt.xlabel('Years of Experience') # Memberikan label pada sumbu X
plt.ylabel('Salary') # Memberikan label pada sumbu Y
plt.show() # Menampilkan grafik
```

Multiple Linear Regresion 50 Start Up



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
# Importing the libraries
```

```python
dataset = pd.read_csv('50_Startups.csv') # Membaca dataset dari file CSV
```

```python
X = dataset.iloc[:, :-1].values  # Semua kolom kecuali yang terakhir
y = dataset.iloc[:, -1].values   # Kolom terakhir adalah target
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
# Menerapkan OneHotEncoder pada kolom kategori (kolom ke-3, indeks 2)
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# Avoiding the Dummy Variable Trap
# Menghapus salah satu kolom dari variabel dummy agar tidak terjadi perangkap variabel dummy
X = X[:, 1:]
```

```python
# Splitting the dataset into the Training set and Test set
# Memisahkan dataset menjadi training set dan test set (80% training dan 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Training the Multiple Linear Regression model on the Training set
# Melatih model regresi linear ganda pada data training
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
# Memprediksi hasil dari data test
y_pred = regressor.predict(X_test)
```

```python
# Visualisasi hasil prediksi (Opsional)
for i in range(len(y_test)):
    print(f'Actual: {y_test[i]}, Predicted: {y_pred[i]}')
```

Polynomial Regresion Position Salaries



```python
import numpy as np #import numpy
import matplotlib.pyplot as plt # import matplotlib
import pandas as pd # import pandas
# Importing the libraries
# Mengimpor library yang diperlukan
```

```python
# Importing the dataset
# Membaca dataset dari file CSV
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
# Memisahkan variabel independen (Level) dan dependen (Salary)
X = dataset.iloc[:, 1:-1].values  # Kolom ke-2 (Level) sebagai variabel independen
y = dataset.iloc[:, -1].values    # Kolom terakhir (Salary) sebagai variabel dependen
```

```python
# Training the Linear Regression model on the whole dataset
# Melatih model regresi linear sederhana dengan seluruh dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Training the Polynomial Regression model on the whole dataset
# Melatih model regresi polinomial dengan seluruh dataset
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)  # Degree = 4 untuk membuat fitur polinomial
X_poly = poly_reg.fit_transform(X)  # Mengubah X menjadi fitur polinomial
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)  # Melatih model regresi linier menggunakan fitur polinomial
```

```python
# Visualising the Linear Regression results
# Visualisasi hasil regresi linear
plt.scatter(X, y, color = 'red')  # Plot data aktual dalam bentuk titik merah
plt.plot(X, lin_reg.predict(X), color = 'blue')  # Plot garis regresi linear
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results
# Visualisasi hasil regresi polinomial
plt.scatter(X, y, color = 'red')  # Plot data aktual dalam bentuk titik merah
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')  # Plot garis regresi polinomial
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Polynomial Regression results (for higher resolution and smoother curve)
# Visualisasi hasil regresi polinomial dengan resolusi lebih tinggi
X_grid = np.arange(min(X), max(X), 0.1)  # Meningkatkan resolusi plot
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')  # Plot data aktual dalam bentuk titik merah
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')  # Plot garis regresi polinomial yang halus
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predicting a new result with Linear Regression
# Memprediksi hasil baru dengan regresi linear
lin_reg.predict([[6.5]])
```

```python
# Predicting a new result with Polynomial Regression
# Memprediksi hasil baru dengan regresi polinomial
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
